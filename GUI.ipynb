```python
import tkinter as tk #Logic 
from tkinter import ttk #Widgets
 # UDF
def ctof():
    x = inputVar.get()
    y = x * 9/5 + 32
    outputVar.set((str(round(y, 2)) + " °F"))
window =  tk.Tk()
window.title("Temperature Converter")
window.geometry("400x300")
 #1
 # create a widget with change font size
heading = ttk.Label(master=window, 
                    text="°C (Celsius) to °F (Fahrenheit) Converter", 
                    font="calibri 14 bold",
                    background="#4A90E2",
                    foreground="white",
                    padding=10)
inputVar = tk.IntVar()
 #packing
heading.pack()

 #3
 #hidden frame
inputFrame = ttk.Frame(master=window)
 #2
 # create another widget
inputBox = ttk.Entry(master=inputFrame, textvariable=inputVar)
button = tk.Button(master=inputFrame,   #bind on it inputvar
                   text="Convert",
                   bg = "#F5A623",
                   fg = "white",
                   font=("Calibri", 11, "bold"),
                   command=ctof) #add function
 #packing
inputBox.pack(side="left", padx=10) #add some gape on x-axis 10 pixle
button.pack(side="left") 
inputFrame.pack(pady=20) #add some gape on hidden frame on y-axis 10 pixle
outputVar = tk.StringVar()
 #result
output =  ttk.Label(master=window, 
                    text="Calculated value", 
                    textvariable=outputVar,
                    font=("Calibri", 14, "bold")
                   )#bind & overwrite previous text
output.pack()
window.mainloop()
print("Application End")
```
